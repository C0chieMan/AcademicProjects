```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu


# Read the CSV files into DataFrames
df1 = pd.read_csv('sequential_times.csv', header=None, names=['Data'])
df2 = pd.read_csv('paralelN4_times.csv', header=None, names=['Data'])
df3 = pd.read_csv('paralelN8_times.csv', header=None, names=['Data'])
df4 = pd.read_csv('paralelN12_times.csv', header=None, names=['Data'])
df5 = pd.read_csv('paralelN4NoPhaser_times.csv', header=None, names=['Data'])
df6 = pd.read_csv('paralelN8NoPhaser_times.csv', header=None, names=['Data'])
df7 = pd.read_csv('paralelN12NoPhaser_times.csv', header=None, names=['Data'])


data1 = df1.iloc[:30, :]
data2 = df2.iloc[:30, :]
data3 = df3.iloc[:30, :]
data4 = df4.iloc[:30, :]
data5 = df5.iloc[:30, :]
data6 = df6.iloc[:30, :]
data7 = df7.iloc[:30, :]

statistic, p_value = mannwhitneyu(data1, data3)
print("\nMann-Whitney U Test between Sequencial and the version with the best time (8 cores and phaser):")
print(f"P-value: {p_value}")

last_row1 = df1.iloc[-1, 0]
last_row2 = df2.iloc[-1, 0]
last_row3 = df3.iloc[-1, 0]
last_row4 = df4.iloc[-1, 0]
last_row5 = df5.iloc[-1, 0]
last_row6 = df6.iloc[-1, 0]
last_row7 = df7.iloc[-1, 0]


plt.boxplot([data1['Data'], data2['Data'], data3['Data'], data4['Data'], data5['Data'], data6['Data'], data7['Data']], labels=['seq','N4', 'N8', 'N12','N4NP','N8NP','N12NP'])
plt.xlabel('Dataset')
plt.ylabel('Time in Nano')
plt.title('Box Plot of all the program runs')
plt.show()

print("Sequential in seconds: ", (last_row1/1e9)/60)
print("Paralel w/ 4 cores and phahser: " , (last_row2/1e9)/60)
print("Paralel w/ 8 cores and phahser: " , (last_row3/1e9)/60)
print("Paralel w/ 12 cores and phahser: ", (last_row4/1e9)/60)
print("Paralel w/ 4 cores and  no phahser: ", (last_row5/1e9)/60)
print("Paralel w/ 8 cores and no phahser: ", (last_row6/1e9)/60)
print("Paralel w/ 12 cores and no phahser: ", (last_row7/1e9)/60)
```
